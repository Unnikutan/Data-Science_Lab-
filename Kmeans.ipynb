```python
from sklearn import datasets
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
x = iris.data[:,:2]
y = iris.target
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap="prism")
plt.xlabel("Sepal Length",fontsize=16)
plt.ylabel("Sepel Width",fontsize=16)
plt.show()
```

```python
km = KMeans(n_clusters=3,init="k-means++",n_init=10,verbose=0,tol=0.0001,algorithm="auto",random_state=21,max_iter=300,copy_x="True")
km.fit(x)
center = km.cluster_centers_
print(center)
lbl = km.labels_
print(lbl)
```

```python
fig, axes = plt.subplots(1,2,figsize=(16,8))
axes[0].scatter(x[:,0],x[:,1],c=y,cmap="prism",edgecolor="k",s=85)
axes[1].scatter(x[:,0],x[:,1],c=lbl,cmap="jet",edgecolor="k",s=85)
axes[0].set_xlabel("Sepal Length",fontsize=16)
axes[0].set_ylabel("Sepal Width",fontsize=16)
axes[1].set_xlabel("Sepal Length",fontsize=16)
axes[1].set_ylabel("Sepal Width",fontsize=16)
axes[0].tick_params(direction="in",length=10,width=5,color="blue",labelsize=16)
axes[1].tick_params(direction="out",length=10,width=5,color="black",labelsize=16)
axes[0].set_title("Actual",fontsize=20)
axes[1].set_title("Predicted",fontsize=20)
plt.show()
```
